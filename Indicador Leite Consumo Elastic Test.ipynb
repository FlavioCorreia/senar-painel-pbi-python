```python
import pandas as pd
import sqlalchemy
import datetime
pd.set_option('display.max_columns', 200)
from IPython.display import display
import os
from elasticsearch import Elasticsearch
from elasticsearch import RequestsHttpConnection
```

```python
query_all = {
    "query": {
        "match_all": {}
    }
}
```

```python
host = 'api-corp.cna.org.br'
port = '443'
usuario = 'asper'
senha = 'Asper@TI#'
```

```python
es = Elasticsearch([f'http://{host}:{port}'],
                  connection_class=RequestsHttpConnection, 
                   http_auth=(usuario, senha), 
                   use_ssl=True, verify_certs=False)
es
```

```python
es.indices.get_alias(index="*")
```

```python
es.indices.get_alias(index="sisateg_rel_indicador_leite")
```

```python
index='sisateg_rel_indicador_leite'
scroll_size=5000
scroll_time='15m'
query={
    "query": {
        "match_all": {}
    }
}
```

```python
scroll = es.search(
            index=index, 
            #size=scroll_size, 
            scroll=scroll_time, 
            body=query)

scroll_id = scroll['_scroll_id']
round_num = 1
```
